## Coupled Harmoic Oscilators
solving as an eigenvalue equation

eq1)  $ m\ddot{x_1} = -2kx_1 + kx_2 $  
eq2)  $ m\ddot{x_2} =  kx_1 - 2kx_2 $  
eq3)  $ x_1 = A_1 e^{-i\omega t} $  
eq4)  $ x_1 = A_1 e^{-i\omega t} $  
combining the equations produces  
eq5)  $ -m \omega^2 A_1 = -2kA_1 +kA_2 $  
eq6)  $ -m \omega^2 A_2 = -2kA_2 +kA_1 $  
  
$ MA = - \omega^2 A $ where $ M =\begin{pmatrix} -\frac{2k}{m} & \frac{k}{m} \\ \frac{k}{m}  & -\frac{2k}{m} \end{pmatrix}  $ and $ A = \begin{pmatrix} A_1 \\ A_2 \end{pmatrix} $  
thus eigenvalues are $ -\omega^2 $

# 2x2

```python
def two_by_two(m):

    comparison = np.array([100000,100000])
    
    for i in range(100000):
        
        #print(f"A at begging of loop {i} is {matrix}")
        m = np.transpose(m)     #transposing the matrix so that each c vector is its own row as i fond it easier to understand
        #print(f"transpsosed matrix {matrix}")
    
        # c1 to f1
        c1 = m[0]
        f = np.array(c1)
        #print(f" f1 {f[0:]}")
        f1 = c1

        # c2 to f2
        c2 = m[1]
        #print(f"c2 {c2}")
        mag_f1_squared = np.square(np.linalg.norm(f1))
        #print(f"mag_f1_square {mag_f1_squared}")
        c2_dot_f1 = np.dot(c2,f1)
        #print(f"c2 dot f1 {c2_dot_f1}")
        f2 = c2 - ((c2_dot_f1 / mag_f1_squared) * f1)
        #print(f"f2 {f2}")
        f = np.vstack((f,f2))
        #print(f"f {f}")

        # Q values
        q1 = f1/np.linalg.norm(f1)
        #print(f"q1 {q1}")
        q2 = f2/np.linalg.norm(f2)
        #print(f"q2 {q2}")
        Q = np.vstack((q1,q2))
        #print(f"Q {Q}")

        # U matrix
        u11 = np.linalg.norm(f1)
        u12 = np.dot(c2,q1)
        u21 = 0
        u22 = np.linalg.norm(f2)
        U = np.array([[u11, u21], [u12, u22]])
        #print(f"U {U}")

        # new A
        Q = np.transpose(Q)     # transpoing the q and u matrix to what they should be for cross product
        U = np.transpose(U)
        #print(f" shape of Q {Q.shape}, shape of U {U.shape}")
        new_A = U@Q
        #print(f"shape of new A {new_A.shape}")
        #print(f"new A matrix {new_A}")
        
        m = new_A

        comp_eigs = np.array([m[0,0], m[1,1]])

        #print(f"compeigs {np.round(comp_eigs,decimals = 7)}, comparison {np.round(comparison, decimals = 7)}")
        if (np.round(comp_eigs,decimals = 7) == np.round(comparison, decimals = 7)).all():
            #print(f"i: {i}")
            break
        else:
            pass

        comparison = comp_eigs

    if i == 99999:
        raise StopIteration("The calculations are not converging")
    else:
        pass
        
    eigvals = np.round(np.array([m[0,0], m[1,1]]), decimals = 7)
    #print(f"eigenvalues {eigvals}")
    return eigvals
```

# 3x3

```python
def three_by_three(m):

    comparison = np.array([100000,100000,100000])
    for i in range(100000):
    
        #print(f"A at begging of loop {i} is {m}")
        m = np.transpose(m)     #transposing the matrix so that each c vector is its own row as i fond it easier to understand
        #print(f"transpsosed matrix {m}")
    
        # c1 to f1
        c1 = m[0]
        f = np.array(c1)
        #print(f" f1 {f}")
        f1 = c1
    
        # f2
        c2 = m[1]
        #print(f"c2 {c2}")
        
        mag_f1_squared = np.square(np.linalg.norm(f1))
        #print(f"mag_f1_square {mag_f1_squared}")
        c2_dot_f1 = np.dot(c2,f1)
        #print(f"c2 dot f1 {c2_dot_f1}")
        
        f2 = c2 - ((c2_dot_f1 / mag_f1_squared) * f1)
        #print(f"f2 {f2}")
        f = np.vstack((f,f2))
        #print(f"f {f}")
    
        # f3
        c3 = m[2]
        #print(f"c3 {c3}")
        
        mag_f2_squared = np.square(np.linalg.norm(f2))
        #print(f"mag_f2_square {mag_f2_squared}")
        c3_dot_f2 = np.dot(c3,f2)
        #print(f"c3 dot f2 {c3_dot_f2}")
    
        #print(f"mag_f1_square {mag_f1_squared}")
        c3_dot_f1 = np.dot(c3,f1)
        #print(f"c3 dot f1 {c3_dot_f1}")
    
        f3 = c3 - ((c3_dot_f2/mag_f2_squared)*f2) - ((c3_dot_f1/mag_f1_squared)*f1)
        #print(f"f3 {f3}")
        f = np.vstack((f,f3))
        #print(f"f {f}")
    
        # Q values
        q1 = f1/np.linalg.norm(f1)
        #print(f"q1 {q1}")
        q2 = f2/np.linalg.norm(f2)
        #print(f"q2 {q2}")
        q3 = f3/np.linalg.norm(f3)
        #print(f"q3 {q3}")
        Q = np.vstack((q1,q2,q3))
        #print(f"Q {Q}")
    
        # U matrix
        u11 = np.linalg.norm(f1)
        u12 = np.dot(c2,q1)
        u13 = np.dot(c3,q1)
        u21 = 0
        u22 = np.linalg.norm(f2)
        u23 = np.dot(c3,q2)
        u31 = 0
        u32 = 0
        u33 = np.linalg.norm(f3)
        U = np.array([[u11, u21, u31], [u12, u22, u32], [u13, u23, u33]])
        #print(f"U {U}") 
    
        # new A
        Q = np.transpose(Q)     # transpoing the q and u matrix to what they should be for cross product
        U = np.transpose(U)
        #print(f" shape of Q {Q.shape}, shape of U {U.shape}")
        new_A = U@Q
        #print(f"shape of new A {new_A.shape}")
        #print(f"new A matrix {new_A}")
        m = new_A

        comp_eigs = np.array([m[0,0], m[1,1], m[2,2]])
        #print(f"compeigs {np.round(comp_eigs,decimals = 7)}, comparison {np.round(comparison, decimals = 7)}")
        if (np.round(comp_eigs,decimals = 7) == np.round(comparison, decimals = 7)).all():
            #print(f"i: {i}")
            break
        else:
            pass

    
        
        comparison = comp_eigs


    if i == 99999:
        raise StopIteration("The calculations are not converging")
    else:
        pass
        
    eigvals = np.round(np.array([m[0,0], m[1,1], m[2,2]]), decimals = 7)
    #print(f"eigenvalues {eigvals}")
    return eigvals
```

## 2x2 Eigenvalue check
using the characteristic equation. $ det(A -\lambda I) = 0	$  
$ det\biggl({ \begin{pmatrix}A_{11} & A_{12} \\A_{21} & A_{22}\end{pmatrix} - \begin{pmatrix}\lambda & 0 \\ 0 & \lambda \end{pmatrix} }\biggr) = 0 $  
$ det\begin{pmatrix}A_{11} - \lambda & A_{12} \\A_{21} & A_{22} - \lambda \end{pmatrix} = 0 $  
$ ( A_{11} - \lambda)( A_{22} -\lambda) - A_{12}A_{21} = 0 $  
$ 1\lambda^2 + (-A_{11}-A_{22})\lambda + (A_{11}A_{22}-A_{12}A_{21}) = 0 $ 
Using the quadratic formula $ \lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $  
$ a = 1 $  
$ b = -A_{11}-A_{22} $  
$ c = A_{11}A_{22}-A_{12}A_{21} $  
#### $ \lambda  = \frac{-(-A_{11}-A_{22}) +\sqrt{(-A_{11}-A_{22})^2 - 4(1)(A_{11}A_{22} +A_{12}A_{21})}}{2(A_{11}A_{22})} $ and $ \lambda  = \frac{-(-A_{11}-A_{22}) -\sqrt{(-A_{11}-A_{22})^2 - 4(1)(A_{11}A_{22} +A_{12}A_{21})}}{2(A_{11}A_{22})} $

```python
def two_by_two_check(m):
    A11 = m[0, 0]
    A12 = m[0, 1]
    A21 = m[1, 0]
    A22 = m[1, 1]

    #a = A11*A22
    a = 1
    b = - A11 - A22
    c = (A11 * A22) - (A12 * A21)
    #print(f"a={a}, b={b}, c={c}")

    lambda_plus = (-b + np.sqrt(b**2 -4*a*c )) / (2*a)
    lambda_minus = (-b - np.sqrt(b**2 -4*a*c )) / (2*a)

    check_eigs = np.array([lambda_plus, lambda_minus])

    return check_eigs
```

## Determinant Check

```python
def det_check(ma, skip = False):
    """Calcaulting the determinate of a 2x2 or 3x3 matrix.
    If the deterinant is zero it stops the code by produceing a custom error.
    The function can also tell the code to skip a matrix if its determinant = 0, so that if calculating a number of matrices a specific one doesnt hold up the code.
    skip = True then it skips. skip = False then it continues"""
    if ma.shape == (2,2):
        lead = ma[0,0] * ma[1,1]
        trail = ma[0,1] * ma[1,0]
        det = lead - trail

    elif ma.shape == (3,3):
        print("3x3")
        sub_det1 = (ma[1,1] * ma[2,2]) - (ma[1,2] * ma[2,1])
        sub_det2 = (ma[1,0] * ma[2,2]) - (ma[1,2] * ma[2,0])
        sub_det3 = (ma[1,0] * ma[2,1]) - (ma[1,1] * ma[2,0])
        #print(sub_det1, sub_det2, sub_det3)
        
        det = (ma[0,0]*sub_det1) - (ma[0,1]*sub_det2) + (ma[0,2]*sub_det3) 
        #print(det)
        
    else:
        raise TypeError("Matrix is not supported by code, either it is too big or it's not square")

        
    if det == 0 and skip == True:
        return "skip"
    
    elif det == 0:
        raise ValueError("the determinent is zero, and thus cannot be calulclated using QU factorisation")
        
    else:
        pass
```

## Calculating frequencies for changing masses in various situations  
Equal masses  
One mass is double the other  
one mass is constant  

```python
def mass_change(M1, M2, K):
    """This function calculates the the frequencies that changing the masses produces.
    it takes in two masses and a spring constant, the two amsses are so that you can change the masses from differenct staring points or the same.
    M1: mass 1
    M2: mass 2
    K: spring constant
    returns four arrays two are the masses, and the other two are the calculated frequencies"""
    frequencies1 = np.array([])
    frequencies2 = np.array([])
    mass1 = np.array([])
    mass2 = np.array([])
    
    for i in range(50):

        matrix =  np.array([[-2*K/M1, K/M1], [K/M2, -2*K/M2]])
        mass1 = np.append(mass1, M1)
        mass2 = np.append(mass2, M2)

        check = det_check(matrix, skip = True)
        if check == "skip":
            print(f"The matrix {matrix} was skipped due to a determinant of 0")
            continue
        else:
            pass          
        
        temp_eigenvalues = two_by_two(matrix)
        temp_omega1 = np.sqrt(-temp_eigenvalues[0])
        temp_omega2 = np.sqrt(-temp_eigenvalues[1])
        frequencies1 = np.append(frequencies1, temp_omega1)
        frequencies2 = np.append(frequencies2, temp_omega2)

        M1 = M1 + 1
        M2 = M1 + 1

    return mass1, mass2, frequencies1, frequencies2
```

```python
def mass_change_double(M1, K):
    """This function calculates the the frequencies that changing the masses produces.
   Takes in one mass and calculates the frequencies for when one mass is twice the mass of the other.
    M1: mass 1
    K: spring constant
    returns four arrays two are the masses, and the other two are the calculated frequencies"""
    frequencies1 = np.array([])
    frequencies2 = np.array([])
    mass1 = np.array([])
    mass2 = np.array([])

    M2 = 2 *M1
    
    for i in range(50):

        matrix =  np.array([[-2*K/M1, K/M1], [K/M2, -2*K/M2]])
        mass1 = np.append(mass1, M1)
        mass2 = np.append(mass2, M2)

        check = det_check(matrix, skip = True)
        if check == "skip":
            print(f"The matrix {matrix} was skipped due to a determinant of 0")
            continue
        else:
            pass   
        
        temp_eigenvalues = two_by_two(matrix)
        temp_omega1 = np.sqrt(-temp_eigenvalues[0])
        temp_omega2 = np.sqrt(-temp_eigenvalues[1])
        frequencies1 = np.append(frequencies1, temp_omega1)
        frequencies2 = np.append(frequencies2, temp_omega2)

        M1 = M1 + 1
        M2 = 2 * M1

    return mass1, mass2, frequencies1, frequencies2
```

```python
def mass_change_single(M1, M2, K):
    """This function calculates the the frequencies that changing  one mass produces.
   Takes in two mass and calculates the frequencies when M1 changes.
    M1: mass 1
    M2: mass 2
    K: spring constant
    returns four arrays two are the masses, and the other two are the calculated frequencies"""
    frequencies1 = np.array([])
    frequencies2 = np.array([])
    mass1 = np.array([])
    mass2 = np.array([])
    
    for i in range(50):

        matrix =  np.array([[-2*K/M1, K/M1], [K/M2, -2*K/M2]])
        mass1 = np.append(mass1, M1)
        mass2 = np.append(mass2, M2)

        check = det_check(matrix, skip = True)
        if check == "skip":
            print(f"The matrix {matrix} was skipped due to a determinant of 0")
            continue
        else:
            pass   
        
        temp_eigenvalues = two_by_two(matrix)
        temp_omega1 = np.sqrt(-temp_eigenvalues[0])
        temp_omega2 = np.sqrt(-temp_eigenvalues[1])
        frequencies1 = np.append(frequencies1, temp_omega1)
        frequencies2 = np.append(frequencies2, temp_omega2)

        M1 = M1 + 1

    return mass1, mass2, frequencies1, frequencies2
```
