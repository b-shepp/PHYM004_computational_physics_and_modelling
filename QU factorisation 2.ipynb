### $ f_1 = c_1 $
### $ f_2 = c_2 - \frac{c_2 \cdot f_1}{||f_1||} f_1 $  
### $ f_3 = c_3 - \frac{c_3 \cdot f_2}{{||f_2||}^2} f_2 - \frac{c_3 \cdot f_1}{{||f_1||}^2} f_1 $
### $ f_4 = c_4 - \frac{c_4 \cdot f_3}{{||f_3||}^2} f_3 - \frac{c_4 \cdot f_2}{{||f_2||}^2} f_2 - \frac{c_4 \cdot f_1}{{||f_1||}^2} f_1 $  

### $ q_k = \frac{f_k}{||f_k||}$  
$U =  \begin{pmatrix}
||f_1|| & c_2 \cdot q_1 \\
0 & ||f_2|| 
\end{pmatrix} $

### 2x2

```python
#getting c1 as f1
for i in range(1000):
    f = np.array(matrix[0])
    #f = np.vstack((f, matrix[1]))
    #print(f"f {f}")

    squared_mag_f1 = np.square(np.linalg.norm(f[0:]))
    #print(f"squared_mag_f1 {squared_mag_f1}")
    c2 = matrix[1]
    whole_term = (np.dot(c2, f[0:]) / squared_mag_f1) * f[0:]
    tot = c2 - whole_term
    #print(f"f2 {tot}")
    f = np.vstack((f,tot))
    #print(f"f {f}")

    #print(f"f[0] {f[0]}")
    #print(f"f[1] {f[1]}")
    q1 = f[0] / (np.linalg.norm(f[0]))
    q2 = f[1] / (np.linalg.norm(f[1]))
    #print(f"q1 {q1} : q2 {q2}")
    q = np.vstack((q1,q2))
    #print(f"q {q}")

    #print(f"q1 {q[0]}")
    c2_q1 = np.dot(c2,q[0])
    #print(f"c2_q1 {c2_q1}")
    U = np.array([[np.linalg.norm(f[0]), c2_q1],[0, np.linalg.norm(f[1])]])
    #print(f"u {U}")
    matrix = np.transpose(np.matmul(U,q))
    #print(f"new A_{i+1} {matrix}")

print(f"eigenvalues {matrix}")
```
