# QU Factorisation Algorithm
Used for solving eigenvalues

```python
import numpy as np
```

```python
def two_by_two(m):

    for i in range(100):

        #print(f"A at begging of loop {i} is {m}")
        m = np.transpose(m)     #transposing the matrix so that each c vector is its own row as i fond it easier to understand
        #print(f"transpsosed matrix {m}")
    
        # c1 to f1
        c1 = m[0]
        f = np.array(c1)
        #print(f" f1 {f[0:]}")
        f1 = c1

        # f2
        c2 = m[1]
        #print(f"c2 {c2}")
        mag_f1_squared = np.square(np.linalg.norm(f1))
        #print(f"mag_f1_square {mag_f1_squared}")
        c2_dot_f1 = np.dot(c2,f1)
        #print(f"c2 dot f1 {c2_dot_f1}")
        f2 = c2 - ((c2_dot_f1 / mag_f1_squared) * f1)
        #print(f"f2 {f2}")
        f = np.vstack((f,f2))
        #print(f"f {f}")

        # Q values
        q1 = f1/np.linalg.norm(f1)
        #print(f"q1 {q1}")
        q2 = f2/np.linalg.norm(f2)
        #print(f"q2 {q2}")
        Q = np.vstack((q1,q2))
        #print(f"Q {Q}")

        # U matrix
        u11 = np.linalg.norm(f1)
        u12 = np.dot(c2,q1)
        u21 = 0
        u22 = np.linalg.norm(f2)
        U = np.array([[u11, u21], [u12, u22]])
        #print(f"U {U}")

        # new A
        Q = np.transpose(Q)     # transpoing the q and u matrix to what they should be for cross product
        U = np.transpose(U)
        #print(f" shape of Q {Q.shape}, shape of U {U.shape}")
        new_A = U@Q
        #print(f"shape of new A {new_A.shape}")
        #print(f"new A matrix {new_A}")
        
        m = new_A

    eigvals = np.array([m[0,0], m[1,1]])
    #print(f"eigenvalues {eigvals}")
    return eigvals
```

### $ f_1 = c_1 $
### $ f_2 = c_2 - \frac{c_2 \cdot f_1}{||f_1||} f_1 $  
### $ f_3 = c_3 - \frac{c_3 \cdot f_2}{{||f_2||}^2} f_2 - \frac{c_3 \cdot f_1}{{||f_1||}^2} f_1 $
### $ f_4 = c_4 - \frac{c_4 \cdot f_3}{{||f_3||}^2} f_3 - \frac{c_4 \cdot f_2}{{||f_2||}^2} f_2 - \frac{c_4 \cdot f_1}{{||f_1||}^2} f_1 $  

### $ q_k = \frac{f_k}{||f_k||}$  
$U =  \begin{pmatrix}
||f_1|| & c_2 \cdot q_1 \\
0 & ||f_2|| 
\end{pmatrix} $  
### $ A_2 = U_1 Q_1 $
