# Boundary problems

```python
import numpy as np
```

### Shooting method  
Secant method  
$ y(t_1,s)-h=0 $  
s = initial vellocity

```python
def Y(t,s):
    y = t**2 *s
    return y
```

```python
def Secant_method(guesses, k_range,t_1,h_):
    
    for i in range(k_range):
        s_k1 = guesses[-1]
        s_k2 = guesses[-2]
        num = (Y(t_1,s_k1)-h_)*(s_k1-s_k2)
        den = Y(t_1,s_k1) - Y(t_1,s_k2)
        s_k = s_k1-(num/den)
        guesses = np.append(guesses,s_k)

    return guesses[-1]
```

```python
initial_guesses = np.array([8,12])
k = 3
t1 = 10
h=20
print(Secant_method(initial_guesses, k, t1,h))
```
