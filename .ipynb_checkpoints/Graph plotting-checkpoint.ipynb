```python
import matplotlib.pyplot as plt
```

```python
def plot(x_data, y_data, title, log_x, log_y, x_label, y_label, x_limit, y_limit, legend_label, line_style, colour):
    """Function for plotting a simple scatter plot.
    x_data: the x axis data.
    y_data: the y axis data.
    title: title of the plot.
    log_x: do you want the x axis to be logarithmic. True for yes, False for no.
    log_y: do you want the y axis to be logarithmic. True for yes, False for no.
    x_label: the label for the x axis.
    y_label: the label for the y axis.
    x_limit: the limits for the x axis. as brackets (lower limit, upper limit), or 'False' if they are not wanted.
    y_limit: the limits for the y axis. as brackets (lower limit, upper limit), or 'False' if they are not wanted.
    legend_label: label for the points that will go into the legend.
    line_style: The differnt line styles between the data points, '   ', 'solid', 'dotted', 'dashed', 'dashdot'.
    colour: the colour the data points and line are to be."""
    
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

    
    if legend_label == False:
        print("No legend label")
        ax.plot(x_data, y_data, marker = "x", linestyle = line_style, color = colour)
        
    elif type(legend_label) == type("this is a string"):
        print("legend label is a string")
        ax.plot(x_data, y_data, marker = "x", linestyle = line_style, color = colour, label = legend_label)
                
    else:
        print("error legend label is neither string nor false")

    if x_limit == False and y_limit == False:
        print("No limits")
        pass
        
    elif x_limit == False and y_limit != False:
        print("Only a y limit")
        ax.set_ylim(y_limit)
        
    elif y_limit == False and x_limit != False:
        print("Only a x limit")
        ax.set_xlim(x_limit)
        
    elif y_limit != False and x_limit != False:
        print("x and y limits")

    else:
        print("check the axis limits")


    
    if log_x == True:
        print("logarithmic x axis")
        ax.set_xscale("log")
        
    elif log_x == False:
        print("linear x axis")
        pass
    else:
        print("log_x has not been properly filed. 'True' for a log axis, 'False for regular'.")

    if log_y == True:
        print("logarithmic y axis")
        ax.set_yscale("log")
    elif log_y == False:
        print("linear y axis")
        pass
    else:
        print("log_y has not been properly filed. 'True' for a log axis, 'False for regular'.")

    
    return fig
    
```
